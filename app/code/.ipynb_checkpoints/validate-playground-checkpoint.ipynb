```python
import json
import urllib3
import certifi
def webSearch(sn, ver):
    ''' Searches the GESDISC Data Services API endpoint with given parameters:
        sn/shortname e.g. 'SNDRSNIML2CCPRET or M2T1NXSLV')
        ver/version (collection version) e.g. '2' or '5.12.4'
        Returns: response from request
    '''
    ## Pool Manager instance to create a request
    http = urllib3.PoolManager(
        cert_reqs='CERT_REQUIRED', ca_certs=certifi.where())
    ## GESDISC API endpoint
    url = 'https://disc.gsfc.nasa.gov/service/datasets/jsonwsp'
    sn_ver = sn + '_' + ver
    ## Construct search request
    search_request = {'methodname': 'search',
                      'type': 'jsonwsp/request',
                      'version': '1.0',
                      'args': {'search': sn_ver}
                      }

    ## Submit the search request to the GES DISC server
    try:
        r = http.request('POST', url, body=json.dumps(search_request),
                         headers={'Content-Type': 'application/json',
                                  'Accept': 'application/json'})
        response = json.loads(r.data)
        if response['result']['totalResults'] > 1:
            print('Too many items match your search entry')
            print('Try again with exact collection ShortName and version')

        elif response['result']['totalResults'] == 0:
            print('No items were found')
        ## Return one response from request
        else:
            return response

    except KeyError:
        if response['type'] == 'jsonwsp/fault':
            print('ERROR! Faulty request. Please try again.')

def checkServices(response):
    ''' Check if a collection has subsetting services
        and then return variables associated with that specific service.
        Checks for L2 or L34RS or both subsetting services.
        Returns varNameD a dict with keys: L2, L34RS, or both with a 
        list of variable names as their respective values
        Parameter: response from requests
        Returns: varNameD (nested list in dict of variable names)
    '''
    try:  ## Check that variable subsetting services exist
        varNameD = {}
        for ssD in response['result']['items'][0]['services']['subset']:
            if 'variables' in ssD['capabilities'] and 'dataFields' in ssD:
                ## Check what subsetting service it is and add it to dict
                if ssD['agentConfig']['agentId'] == 'SUBSET_LEVEL2' or \
                        'L2' in ssD['agentConfig']['agentId']:
                    varNameD['L2'] = [var['value'].replace('/', '', 1)
                                      for var in ssD['dataFields']]

                elif 'L34RS' in ssD['agentConfig']['agentId']:
                    varNameD['L34RS'] = [var['value'].replace('/', '', 1)
                                         for var in ssD['dataFields']]

        return varNameD

    except KeyError:
        print('Variable subsetting is not available for %s'
              % response['result']['items'][0]['dataset']['id'])
```

```python
short_name = 'TRPSYL2HDOCRS1FS'
ver_number = '1'
num_granules = 'all' #'all' # specify a number or 'all' to check whole collection
```

```python
api_granule = GranuleQuery(mode=CMR_OPS)
#api_granule.mode(mode)
api_granule.parameters(
    short_name= short_name,
    version = ver_number
    #temporal = (datetime(2022, 10, 1, 0, 0, 0), datetime(2022, 10, 13, 0, 0, 0))
    )
#api_granule.mode('')
if num_granules == 'all':
    granules = api_granule.get_all()
else:
    granules = api_granule.get(num_granules)


checked_files = pd.DataFrame(columns = ['short_name', 'version', 'producer_granule_id','native_id'
                                              'default_url','opendap_url','s3_url','updated_date' ])

out_logs = []

for i in range(0,np.shape(granules)[0]):
    producer_granule_id = granules[i]['producer_granule_id']
    native_id = granules[i]['id']
    updated_date = granules[i]['updated']

    default_url = False 
    opendap_url = False
    s3_url = False
    for element in granules[i]['links']:
        if element['rel'] == 'http://esipfed.org/ns/fedsearch/1.1/data#':
            temp = (element['href'].split('.')[-1])
            if temp == 'he5':
                print('default data link: '+ element['href'])
                default_url = element['href']
            if temp == 'nc':
                print('default data link: '+ element['href'])
                default_url = element['href']
            if temp == 'ncml':
                print('default data link: '+ element['href'])
                default_url = element['href']

        elif element['rel'] == 'http://esipfed.org/ns/fedsearch/1.1/service#':
            temp = (element['href'].split('.')[-1])
            if temp == 'he5':
                print('opendap link '+ element['href'])
                opendap_url = element['href']
            if temp == 'nc':
                print('default data link: '+ element['href'])
                opendap_url = element['href']
            if temp == 'ncml':
                print('default data link: '+ element['href'])
                opendap_url = element['href']

        elif element['rel'] == 'http://esipfed.org/ns/fedsearch/1.1/s3#':
            print('s3 link '+ element['href'])
            s3_url = element['href']
    data_out = [short_name,ver_number,producer_granule_id,native_id,default_url,opendap_url,s3_url,updated_date]
    out_logs.append(data_out)

out_logs=np.array(out_logs)

checked_files = pd.DataFrame({'short_name':out_logs[:,0], 'version':out_logs[:,1], 
    'producer_granule_id':out_logs[:,2],'native_id':out_logs[:,3],
    'default_url':out_logs[:,4],'opendap_url':out_logs[:,5],
    's3_url':out_logs[:,6],'updated_date':out_logs[:,7]})

#checked_files = pd.DataFrame({'short_name':out_logs[:,0], 'version':out_logs[:,1], 'producer_granule_id':out_logs[:,2],'native_id':out_logs[:,3], 's3_url':out_logs[:,4],'version_from_s3_link':out_logs[:,5],'versions_match':out_logs[:,6],'granule_within_date_range':out_logs[:,7]})
#checked_files
checked_files.to_csv('prod-checked-files-' + short_name + '-' + ver_number + '.csv')
```
